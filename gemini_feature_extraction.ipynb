```python
import os
import jsonlines
import pandas as pd
import google.generativeai as genai
from tqdm import tqdm
import time
from dotenv import load_dotenv
```

## Step 2: Configure API Key

```python
load_dotenv('/content/drive/MyDrive/SemEval-2024 Task 8/.env')

GEMINI_API_KEY = os.getenv('GEMINI_API_KEY')
genai.configure(api_key=GEMINI_API_KEY)

genai_model = genai.GenerativeModel("gemini-2.0-flash")
```

## Step 3: Define Prompt Template

```python
def build_prompt(text_list):
    joined_texts = "\n\n".join(
        [f"{i+1}. {text}" for i, text in enumerate(text_list)]
    )

    prompt = f"""
You are a linguistic expert. I will give you multiple short texts. For each one, extract the following grammatical and syntactic features:
- Sentence structure (e.g., simple, compound, complex)
- Number of independent and dependent clauses
- Whether it contains participial phrases, hanging participles, or other non-standard structures

Return the result in a CSV format with the following columns:
Index, SentenceStructure, NumIndependentClauses, NumDependentClauses, HasParticipialPhrase, HasHangingParticiple

Here are the texts:
{joined_texts}
"""
    return prompt
```

## Step 4: Define Load Dataset Function

```python
def load_texts_from_jsonl(file_path):
    texts = []
    with jsonlines.open(file_path) as reader:
        for obj in reader:
            texts.append(obj["text"])
    return texts
```

## Step 5: Send Prompts in Batches

```python
def extract_features_batchwise(texts, batch_size=10, sleep_time=2):
    results = []

    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i+batch_size]
        prompt = build_prompt(batch)

        try:
            response = genai_model.generate_content(prompt)
            raw_csv = response.text

            # Convert raw CSV to DataFrame
            from io import StringIO
            df = pd.read_csv(StringIO(raw_csv))
            df["OriginalIndex"] = list(range(i, i+len(batch)))
            results.append(df)

        except Exception as e:
            print(f"Error in batch {i}–{i+batch_size}: {e}")
            continue

        time.sleep(sleep_time)  # be respectful to the API rate

    return pd.concat(results, ignore_index=True)
```

## Step 6: Process and Save CSVs

```python
# File paths
train_file = "/content/drive/MyDrive/SemEval-2024 Task 8/Datasets/subtaskA_train_monolingual.jsonl"
dev_file = "/content/drive/MyDrive/SemEval-2024 Task 8/Datasets/subtaskA_dev_monolingual.jsonl"
test_file = "/content/drive/MyDrive/SemEval-2024 Task 8/Datasets/subtaskA_monolingual.jsonl"

# Load text
train_texts = load_texts_from_jsonl(train_file)
# dev_texts = load_texts_from_jsonl(dev_file)
# test_texts = load_texts_from_jsonl(test_file)

# Extract features
train_features_df = extract_features_batchwise(train_texts, batch_size=10)
# dev_features_df = extract_features_batchwise(dev_texts, batch_size=10)
# test_features_df = extract_features_batchwise(test_texts, batch_size=10)

# Save to CSV
train_features_df.to_csv("/content/drive/MyDrive/SemEval-2024 Task 8/Gemini Features/train_gemini_features.csv", index=False)
# dev_features_df.to_csv("/content/drive/MyDrive/SemEval-2024 Task 8/Gemini Features/dev_gemini_features.csv", index=False)
# test_features_df.to_csv("/content/drive/MyDrive/SemEval-2024 Task 8/Gemini Features/test_gemini_features.csv", index=False)
```
